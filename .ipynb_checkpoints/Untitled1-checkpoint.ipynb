```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10)  # Corrected function name
y = np.arange(0, 10)  # Corrected function name

# Scatter plot
plt.scatter(x, y, c='g')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Scatterplot')

# Save the figure
plt.savefig('Scatterplot.png')

# Line plot with dotted line style
plt.plot(x, y, 'r:', label='Dotted Line')  # Red dotted line

# Show the plots
plt.legend()  # To display the legend
plt.show()
```

```python


x = np.arange(0, 4 * np.pi, 0.1)  # Corrected function name
y = np.sin(x)

plt.plot(x, y, 'r')
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
x = np.arange(0, 5 * np.pi, 0.1)  
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2, 2, 1)
plt.plot(x, y_sin, 'r')
plt.title('Sinewave')

plt.subplot(2, 2, 2)
plt.plot(x, y_cos, 'r')
plt.title('Coswave')
```

```python
x =[2,8, 10]
y = [11, 10, 9]

x2 = [3, 9, 11]
y2 = [6, 15, 17]
plt.bar(x,y)
plt.bar(x2, y2, color = 'g')
plt.title('BarPlot')

plt.show()
```

```python

x = np.random.normal([22, 11, 47, 52, 12, 60, 8, 27, 42])

plt.hist(x)
plt.title('Histogram')
plt.show() 
```

```python
data = [np.random.normal(0, std, 100) for std in range (1, 4)]
plt.title('Boxplopt')
plt.boxplot(data, vert = True, patch_artist = False)
```
