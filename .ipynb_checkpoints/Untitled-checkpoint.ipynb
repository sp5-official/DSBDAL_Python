```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load Titanic dataset
df = sns.load_dataset("titanic")

# Display first few rows
print("Dataset Overview:")
print(df.head())
print("\nDataset Information:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())

# Pairplot - Overall data relationship
sns.pairplot(df)
plt.show()

# Heatmap - Correlation between numerical features
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()

# Category-wise distribution - Survival rate by class
sns.barplot(x="class", y="survived", data=df)
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Class")
plt.show()

# Histogram - Distribution of ticket fare
sns.histplot(df["fare"], bins=30, kde=True)
plt.xlabel("Fare")
plt.ylabel("Count")
plt.title("Distribution of Ticket Fare")
plt.show()

# Histogram - Distribution of fare by class
sns.histplot(df, x="fare", hue="class", bins=30, kde=True)
plt.xlabel("Fare")
plt.ylabel("Count")
plt.title("Distribution of Ticket Fare by Class")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
df_titanic = sns.load_dataset("titanic")

# Save Titanic dataset to CSV
df_titanic.to_csv("titanic_dataset.csv", index=False)

# Display Titanic dataset
print("Titanic Dataset:")
print(df_titanic.head())

# Heatmap - Correlation between numerical features in Titanic dataset
plt.figure(figsize=(10, 6))
sns.heatmap(df_titanic.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap - Titanic Dataset")
plt.show()

# Bar Plot - Categorical Data Visualization for Titanic dataset
plt.figure(figsize=(8, 5))
sns.countplot(x="class", data=df_titanic)
plt.title("Passenger Class Distribution - Titanic Dataset")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()

# Histogram - Numerical Data Distribution for Titanic dataset
plt.figure(figsize=(8, 5))
sns.histplot(df_titanic["fare"], bins=30, kde=True)
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.title("Fare Distribution - Titanic Dataset")
plt.show()
```
