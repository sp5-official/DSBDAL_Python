```python
import pandas as pd
from sklearn.datasets import load_iris
# Load the Iris dataset
iris = load_iris()
# Convert the dataset into a pandas DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)
# Add the target variable 'species' to the DataFrame
df['species'] = iris.target_names[iris.target]
# Display the first few rows to check the dataset
df.head()
```

```python
# Create a histogram for each numeric feature in the dataset
numeric_features = df.drop(columns='species')
plt.figure(figsize=(15, 10))
# Loop over each numeric feature and plot its histogram
for i, feature in enumerate(numeric_features.columns, 1):
 plt.subplot(2, 2, i)
 sns.histplot(df[feature], kde=True, bins=15)
 plt.title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()
```

### Distribution of Sepal and Petal Features (Iris Dataset)
- **Sepal Length and Width:** Sepal width follows a near-normal distribution, while sepal length is more uniformly distributed.
- **Petal Length and Width:** Both petal length and petal width are highly skewed, indicating distinct groupings of species.
- **Insight:** The petal measurements show stronger separation, likely contributing more to species classification.

```python
plt.figure(figsize=(15, 10))
# Loop over each numeric feature and plot its box plot
for i, feature in enumerate(numeric_features.columns, 1):
 plt.subplot(2, 2, i)
 sns.boxplot(x='species', y=feature, data=df)
 plt.title(f'Box plot of {feature}')
plt.tight_layout()
plt.show()
```

## Observations from the box plots:

* **Sepal Length:** *Iris virginica* tends to have the longest sepal length, followed by *Iris versicolor*, and then *Iris setosa*, which has the shortest sepal length. There's some overlap between *versicolor* and *virginica*, but *setosa* is clearly distinct.
* **Sepal Width:** *Iris setosa* generally has the widest sepals. *Iris versicolor* has the narrowest sepal width on average, and *Iris virginica*'s sepal width falls between the other two species with some overlap. *Iris setosa* also shows a few outliers with very narrow sepal widths.
* **Petal Length:** The petal lengths show the most distinct separation between the species. *Iris virginica* has the longest petals, *Iris versicolor* has intermediate petal lengths, and *Iris setosa* has significantly shorter petals. There is minimal overlap between the species in terms of petal length.
* **Petal Width:** Similar to petal length, the petal width also clearly distinguishes the species. *Iris virginica* exhibits the widest petals, *Iris versicolor* has intermediate petal widths, and *Iris setosa* has the narrowest petals. Again, there is little overlap in petal width among the three species.

```python
# Create box plots for each numeric feature
plt.figure(figsize=(12, 6))
for i, column in enumerate(df.select_dtypes(include='number').columns):
 plt.subplot(1, 4, i + 1)
 sns.boxplot(y=df[column], color='lightgreen')
 plt.title(column)
plt.suptitle("Box Plots of Numeric Features - Iris Dataset", fontsize=14)
plt.tight_layout()
plt.show()
```

## Observations of Iris Flower Features

* **Sepal Length (cm):** The sepal length has a median around 5.8 cm. The middle 50% of the data falls between approximately 5.1 cm and 6.4 cm. The distribution appears somewhat symmetric, with a few potential outliers on the higher end.
* **Sepal Width (cm):** The median sepal width is around 3.0 cm. The interquartile range (IQR) is roughly from 2.8 cm to 3.3 cm, indicating a relatively narrow spread for the central 50% of the data. There are several potential outliers, both on the lower and upper ends of the distribution.
* **Petal Length (cm):** The petal length shows a median of approximately 4.4 cm. The IQR extends from about 1.6 cm to 5.1 cm, suggesting a larger spread compared to sepal width. The distribution is noticeably skewed towards lower values.
* **Petal Width (cm):** The petal width has a median around 1.3 cm. The middle 50% of the data lies between roughly 0.3 cm and 1.8 cm, indicating a substantial spread. The distribution is also skewed towards lower values, similar to petal length.
