```python
# Fixing the data issue
np.random.seed(42)
num_samples = 50  # Ensure consistency

data = {
    "Math_Score": np.random.randint(50, 100, num_samples).tolist() + [200, 300],  
    "Reading_Score": np.random.randint(50, 100, num_samples).tolist() + [800, 900],  
    "Writing_Score": np.random.randint(50, 100, num_samples).tolist() + [150, 220],  
    "Placement_Score": np.random.randint(50, 100, num_samples).tolist() + [180, 200],  
}

# Convert to DataFrame (Ensuring all lists are the same length)
df = pd.DataFrame(data)

# Update column names in 'col' list
col = ['Math_Score', 'Reading_Score', 'Writing_Score', 'Placement_Score']

# Boxplot using Pandas
df.boxplot(column=col)
plt.show()

# Boxplot using Seaborn
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, palette="cool", width=0.5)
plt.show()
```

```python

# Compute Z-score for Math_Score
z = np.abs(stats.zscore(df['Math_Score']))  # Use "Math_Score" instead of "MathScore"
# print(z)

threshold = 0.18
sample_outliers = np.where(z < threshold)[0]  # Get indices as a list

# Sort Reading_Score and compute IQR
sorted_rscore = sorted(df['Reading_Score'])
q1 = np.percentile(sorted_rscore, 25)
q3 = np.percentile(sorted_rscore, 75)

IQR = q3 - q1
lwr_bound = q1 - (1.5 * IQR)
upr_bound = q3 + (1.5 * IQR)

print(lwr_bound, upr_bound)

# Detect outliers in Reading_Score
r_outliers = [i for i in sorted_rscore if i < lwr_bound or i > upr_bound]
# print(r_outliers)

# Drop outliers from the DataFrame
new_df = df.drop(index=sample_outliers, inplace=False)  # Avoid modifying df directly
print(new_df)

lwr_bound = q1 - (1.5 * IQR)
upr_bound = q3 + (1.5 * IQR)

print(lwr_bound, upr_bound)

# Detect outliers in Reading_Score
r_outliers = [i for i in sorted_rscore if i < lwr_bound or i > upr_bound]
# print(r_outliers)

# Drop outliers from the DataFrame
new_df = df.drop(index=sample_outliers, inplace=False)  # Avoid modifying df directly
print(new_df)
```
